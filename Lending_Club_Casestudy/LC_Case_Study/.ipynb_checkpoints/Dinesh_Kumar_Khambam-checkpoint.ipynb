```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')


loan = pd.read_csv("loan.csv", encoding = "ISO-8859-1")
```

```python
#To show all the Columns
pd.options.display.max_columns = None
```

```python
#DROPPING ALL COLUMNS HAVING DUPLICATES AND NaN VALUES

##Drop columns with all NaN as they will not be useful for analysis
loan=loan.dropna(axis=1,how='all')


#loan=loan.drop_duplicates()
loan=loan.drop('pub_rec_bankruptcies',axis=1)
loan=loan.drop('chargeoff_within_12_mths',axis=1)
loan=loan.drop('collections_12_mths_ex_med',axis=1)
```

```python
#Below are Customer Behaviour variables and can be dropped as they are not available at the time of loan application.
#loan.next_pymnt_d.unique()
#loan.last_pymnt_amnt 
#loan.last_pymnt_d.nunique()
#loan.mths_since_last_delinq
#df.acc_now_delinq.unique()
#df.delinq_amnt.unique()

loan=loan.drop('acc_now_delinq',axis=1)
loan=loan.drop('delinq_amnt',axis=1)

loan=loan.drop('mths_since_last_delinq',axis=1)
loan=loan.drop('next_pymnt_d',axis=1)
loan=loan.drop('last_pymnt_amnt',axis=1)
loan=loan.drop('last_pymnt_d',axis=1)
loan=loan.drop('tax_liens',axis=1)

loan=loan.drop('out_prncp',axis=1)
loan=loan.drop('out_prncp_inv',axis=1)

loan.info()
```

```python
#-->For below Varabiles there is only Only 1 distinct unique value can be deleted(kind of redundant data columns).   
loan=loan.drop('policy_code',axis=1)
loan=loan.drop('application_type',axis=1)
```

```python
loan.columns
```

```python
#Below are few other columns that cannot be of help for the analysis
#df.url.tolist()  --> Unwanted Columns and can be deleted
#df.total_rec_int.nunique()  --> Total Interest received is not useful  and can be dropped
#df.total_rec_prncp.nunique()    -->Total Principal received till date and can be dropped.
#df.zip_code.nunique()   --> This Demographic varaible not useful for analaysis                   
#df.addr_state.unique() --> Demographic varaible not useful for analaysis
##df.title is more detailed explanation of the purpose we can skip
#df.desc  -->Borrower desciption of purpose ..redendant
#df.issue_d  -->may not be an important field

loan=loan.drop('url',axis=1)
loan=loan.drop('desc',axis=1)
loan=loan.drop('issue_d',axis=1)
loan=loan.drop('zip_code',axis=1)
loan=loan.drop('addr_state',axis=1)
loan=loan.drop('total_rec_int',axis=1)
loan=loan.drop('total_rec_prncp',axis=1)
loan=loan.drop('title',axis=1)
```

```python
#Trim the Dataframe of  Unwanted leading and trailing withspaces
loan= loan.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
loan.loan_status.unique()  
loan=loan.loc[loan['loan_status'].isin(['Fully Paid','Charged Off'])]
#['Fully Paid', 'Charged Off']
#The Loans marked  as 'current' are neither fully paid not defaulted, so we can get rid of get rid of all the current loans
```

# UniVariate ,Univariate Segmented  and Bivariate Analyis of different variabales (Categorical and Continuous)

```python
#NEW DATAFRAMES FOR EACH LOAN STATUS
charged_off=loan.loc[loan['loan_status'].isin(['Charged Off'])]

fully_paid=loan.loc[loan['loan_status'].isin(['Fully Paid'])]
#current=loan.loc[loan['loan_status'].isin(['Current'])]

#PLOT 1:HOME OWNERSHIP AGAINST LOAN STATUS
plot=sns.countplot(charged_off.home_ownership,hue=charged_off["loan_status"])
plt.xlabel("HOME OWNERSHIP")
plt.ylabel("NUMBER OF CHARGED OFF RECORDS")
plt.show()
```

Customers who's home ownership is Rent has the highest number of Charged off accounts.

```python
#PLOT 2: SUBPLOTS SHOWING VERIFICATION STATUS FOR EACH LOAN STATUS
fig=plt.figure(figsize=(12,6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
aax=fig.add_subplot(131)
sns.countplot(x="loan_status",hue='verification_status',data=charged_off)
plt.title("CHARGED OFF")
plt.xlabel("LOAN STATUS")
plt.ylabel("NUMBER OF CHARGED OFF ACCOUNTS")
aay=fig.add_subplot(132)
sns.countplot(x="loan_status",hue='verification_status',data=fully_paid)
plt.title("FULLY PAID")
plt.xlabel("LOAN STATUS")
plt.ylabel("NUMBER OF FULLY PAID ACCOUNTS")
```

There is no Significant Impact of Not-Verified accounts on Accounts getting Charged off.Which means that Not-Verified accounts are not the Significant reason for accounts getting charged off.

```python
#PLOT 3:NUMBER OF CHARGED OFF ACCOUNTS FOR EACH PURPOSE
fig=plt.figure(figsize=(16,16))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
sns.countplot(x="loan_status",hue='purpose',data=charged_off)
plt.xlabel("LOAN STATUS")
plt.ylabel("NUMBER OF CHARGED OFF ACCOUNTS")
```

Loans grated for Purpose of  debt_consolidation has the highest number of Charged off accounts.

```python
#PERFORMING SOME TRANSFORMATIONS ON 'INT_RATE' AND 'EMP_LENGTH' 
loan['int_rate'] = loan['int_rate'].replace(regex='%', value='')
loan['emp_length']=loan['emp_length'].replace(regex=' year',value='')
loan['emp_length']=loan['emp_length'].replace(regex='s',value='')
```

```python
#SPLITTING 'INT_RATE' INTO CATEGORIES AND MAKING A NEW VARIABLE CALLED 'SLOT'
loan['int_rate']=pd.to_numeric(loan['int_rate']).astype(float)
#loan['int_rate']
loan.loc[(loan.int_rate >=5) &
        (loan.int_rate <10),'slot']='5-10'

loan.loc[(loan.int_rate >=10)&
        (loan.int_rate <15) ,'slot']='10-15'
loan.loc[(loan.int_rate >=15)&
        (loan.int_rate <20),'slot']='15-20'
loan.loc[(loan.int_rate >=20)&
        (loan.int_rate <=25),'slot']='20-25'
#loan
```

```python
#PLOT 6: HEATMAP SHOWING AVERAGE ROI WITH RESPECT TO PURPOSE.
pivot1=loan.pivot_table(values='int_rate',index=['loan_status'],columns=['purpose'],aggfunc='mean')
sns.heatmap(pivot1,annot=True)
plt.title("ROI/PURPOSE")
plt.xlabel("PURPOSE AS TOLD BY THE BORROWER")
plt.ylabel("LOAN STATUS")

## fix for mpl bug that cuts off top/bottom of seaborn  ( Reference Link : https://github.com/mwaskom/seaborn/issues/1773)
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show()
```

```python
HeatMap above shows that for  Purpose of  House , Debt Cosolidation and small_business has the highest average Interest rates
a.) for House its the Highest 
b.) for Small business and Debt Consolidation is the second highest.

For House and Small Business there may be collateral to recover which makes it less risky but for Debt Cosolidaiton
there may not be any collateral which makes it very risky.
```

```python
# fill missing values with mean column values for dti
charged_off['dti']=charged_off['dti'].fillna(loan['dti'].mean())

#loan.dti.describe()

charged_off.info()
```

```python
# Derived variable  dti_slot from Debt to Income Ratio(DTI)
charged_off['dti_slot'] = pd.cut(charged_off['dti'], [0,5,10,15,20,25,30], labels=['0-5','5-10','10-15','15-20','20-25','25-30'])
charged_off.dti_slot.describe()
```

```python
#loan.dti
#sns.barplot(loan['dti_slot'])
#plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='dti_slot', y='int_rate',hue='grade',data=charged_off, estimator=np.mean)
plt.title("DTI Buckets Vs Interest Rate")
plt.show()


fig=plt.figure(figsize=(12,12))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
sns.catplot(x='dti_slot',hue='grade',data=charged_off,kind='count')
plt.title("DTI")
plt.xlabel("")
plt.show()
```

Despite having Better DTI the loans with Higher grades still have High risk and their interest rates are Comparatively higher.

```python
#BoxPlot Showing how the Sub_grades affect Interest rates.
loan.boxplot(column='int_rate', by='sub_grade',figsize=(10,6))
plt.show()
```

Higher the Sub-Grade then higher is the Risk and so higher is the Interest Rate.
